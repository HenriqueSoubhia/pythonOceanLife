```python
import pandas as pd

dadosBase = pd.read_excel("dados-zona-morta.xls")

dadosBase
```

```python
def renameHeader(data):
  map = {}

  for col in data.columns:
      map[col] = col.lower().rstrip()

  return data.rename(columns=map)

dadosRenamed = renameHeader(dadosBase)

dadosRenamed
```

```python
dadosFiltered = pd.DataFrame(dadosRenamed, columns=['country', 'system', 'lat','long','comment'])

dadosFiltered

```

```python
from deep_translator import GoogleTranslator

def translateCol(database,colName):
    coluna_traduzida = []

    for index, row in database.iterrows():
        traducao = GoogleTranslator(source='en', target='portuguese').translate(row[colName])
        coluna_traduzida.append(traducao)

        print(index + 1,"/",len(database),"-", colName)

    return coluna_traduzida


paises_traduzidos = translateCol(dadosFiltered,'country')
sistemas_traduzidos = translateCol(dadosFiltered,'system')
comentarios_traduzidos = translateCol(dadosFiltered,'comment')

dadosFiltered
```

```python
databaseA = dadosFiltered
```

```python

dadosFiltered['country'] = paises_traduzidos
dadosFiltered['system'] = sistemas_traduzidos
dadosFiltered['comment'] = comentarios_traduzidos

map = {
    "country":"pais",
    "system":"sistema",
    "comment":"comentario",
}

dadosTraduzidos = dadosFiltered.rename(columns=map)


dadosTraduzidos

```

```python
def convertToJson(database):
  json = database.to_json(orient="records", date_format='iso')
  return json

dadosJson = convertToJson(dadosTraduzidos)

dadosJson
```

```python
def downloadJson(json):
  f = open("newData.json", "w")
  f.write(json)
  f.close()

downloadJson(dadosJson)
```
